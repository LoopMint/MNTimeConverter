```python
from IPython.display import HTML, display

html_code = """
<!DOCTYPE html>
<html>
  <head>
    <meta charset="UTF-8">
    <title>Singapore and Minnesota Clocks</title>
    <style>
      body {
        font-family: Arial, sans-serif;
        font-size: 28px;
        text-align: center;
        padding-top: 60px;
        background-color: #f4f4f4;
      }
      #sg-clock, #mn-clock {
        margin: 20px 0;
        color: #333;
      }
    </style>
    <script>
      function updateClocks() {
        const sgOptions = {
          timeZone: 'Asia/Singapore',
          weekday: 'long',
          year: 'numeric',
          month: 'short',
          day: '2-digit',
          hour: 'numeric',
          minute: '2-digit',
          second: '2-digit',
          hour12: true
        };

        const mnOptions = {
          timeZone: 'America/Chicago',
          weekday: 'long',
          year: 'numeric',
          month: 'short',
          day: '2-digit',
          hour: 'numeric',
          minute: '2-digit',
          second: '2-digit',
          hour12: true
        };

        const sgTime = new Date().toLocaleString('en-US', sgOptions);
        const mnTime = new Date().toLocaleString('en-US', mnOptions);

        document.getElementById("sg-clock").innerHTML = "Singapore: " + sgTime;
        document.getElementById("mn-clock").innerHTML = "Minnesota: " + mnTime;
      }

      setInterval(updateClocks, 1000);
    </script>
  </head>
  <body onload="updateClocks()">
    <div id="sg-clock">Loading Singapore time...</div>
    <div id="mn-clock">Loading Minnesota time...</div>
  </body>
</html>
"""

display(HTML(html_code))
```
